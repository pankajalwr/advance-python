```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/covid_toy.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
si = SimpleImputer()
```

```python
df['fever'] = si.fit_transform(df[['fever']])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb =LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df.head(3)
```

```python
x = df.drop(columns = ['has_covid'])
y = df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
x_train.head()
```

```python
np.round(x_train.describe(), 2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_mn , columns= x_train.columns)
```

```python
np.round(x_train_new.describe() ,2)
```

```python
df2 = pd.read_csv("/content/tips.csv")
```

```python
df2.head(3)
```

```python
df2.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df2.head(3)
```

```python
df2['sex'] = lb.fit_transform(df2['sex'])
df2['smoker'] = lb.fit_transform(df2['smoker'])
df2['day'] = lb.fit_transform(df2['day'])
df2['time'] = lb.fit_transform(df2['time'])
```

```python
x = df2.drop(columns=['total_bill'])
y = df2['total_bill']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state=42)
```

```python
df2.head(3)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sd = StandardScaler()
```

```python
x_train_sd = sd.fit_transform(x_train)
```

```python
x_train_new2 = pd.DataFrame(x_train_sd , columns= x_train.columns)
```

```python
np.round(x_train_new2.describe(), 2)
```
